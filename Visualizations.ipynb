```python
# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Step 2: Load the existing data
df = pd.read_csv('data/shipping_data.csv')

# Step 3: Perform Chi-Square Test
# Contingency table for 'Service' and 'Was package delivered'
contingency_table = pd.crosstab(df['Service'], df['Was package delivered'])

# Perform Chi-Square Test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Step 4: Calculate averages and cost savings
avg_ground = df[df['Service'] == "UPS Ground"]['Cost to deliver'].mean()
avg_surepost = df[df['Service'] == "UPS SurePost"]['Cost to deliver'].mean()
cost_savings = (avg_ground - avg_surepost) * len(df)

# Step 5: Print results
print("=== Contingency Table ===\n", contingency_table, "\n")
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Average Cost - UPS Ground: ${avg_ground:.2f}")
print(f"Average Cost - UPS SurePost: ${avg_surepost:.2f}")
print(f"Estimated Cost Savings for {len(df)} shipments: ${cost_savings:,.2f}")

# Step 6: Visualizations
# Distribution plot of shipping costs
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Cost to deliver', hue='Service', kde=True, bins=50, palette="Set2")
plt.title('Distribution of Shipping Costs by Service')
plt.xlabel('Shipping Cost')
plt.ylabel('Frequency')
plt.legend(title='Shipping Service')
plt.show()

# Boxplot of shipping costs by service
plt.figure(figsize=(10, 6))
sns.boxplot(x='Service', y='Cost to deliver', data=df, palette="Set2")
plt.title('Boxplot of Shipping Costs by Service')
plt.xlabel('Shipping Service')
plt.ylabel('Shipping Cost')
plt.show()
```

```python
# Step 1: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Step 2: Load the existing data
df = pd.read_csv('data/shipping_data.csv')

# Step 3: Perform Chi-Square Test
# Contingency table for 'Service' and 'Was package delivered'
contingency_table = pd.crosstab(df['Service'], df['Was package delivered'])

# Perform Chi-Square Test
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Step 4: Calculate averages and cost savings
avg_ground = df[df['Service'] == "UPS Ground"]['Cost to deliver'].mean()
avg_surepost = df[df['Service'] == "UPS SurePost"]['Cost to deliver'].mean()
cost_savings = (avg_ground - avg_surepost) * len(df)

# Step 5: Print results
print("=== Contingency Table ===\n", contingency_table, "\n")
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Average Cost - UPS Ground: ${avg_ground:.2f}")
print(f"Average Cost - UPS SurePost: ${avg_surepost:.2f}")
print(f"Estimated Cost Savings for {len(df)} shipments: ${cost_savings:,.2f}")

# Step 6: Visualizations
# Distribution plot of shipping costs
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Cost to deliver', hue='Service', kde=True, bins=50, palette="Set2")
plt.title('Distribution of Shipping Costs by Service')
plt.xlabel('Shipping Cost')
plt.ylabel('Frequency')
plt.legend(title='Shipping Service')
plt.show()

# Boxplot of shipping costs by service
plt.figure(figsize=(10, 6))
sns.boxplot(x='Service', y='Cost to deliver', data=df, palette="Set2")
plt.title('Boxplot of Shipping Costs by Service')
plt.xlabel('Shipping Service')
plt.ylabel('Shipping Cost')
plt.show()

# Step 7: Failure Rate Visualizations
# Calculate the failure rate (1 - success rate) for each service
failure_rate = df.groupby('Service')['Was package delivered'].apply(lambda x: 1 - x.mean())

# Plot the failure rate
plt.figure(figsize=(8, 5))
sns.barplot(x=failure_rate.index, y=failure_rate.values, palette="Set2")
plt.title('Failure Rate of UPS Ground vs UPS SurePost')
plt.xlabel('Shipping Service')
plt.ylabel('Failure Rate')
plt.ylim(0, 1)  # Ensures y-axis is between 0 and 1
plt.show()

# Step 8: Conclusion in Markdown (you can use this in your Jupyter markdown cell)
```

```python
# Plot the failure rates of both services
failure_rate_data = df.groupby('Service')['Was package delivered'].mean()
failure_rate_data = 1 - failure_rate_data  # Convert success rate to failure rate

plt.figure(figsize=(8, 6))
sns.barplot(x=failure_rate_data.index, y=failure_rate_data.values, palette='Blues')

# Adjusting axis
plt.title('Failure Rate by Service', fontsize=16)
plt.xlabel('Shipping Service', fontsize=14)
plt.ylabel('Failure Rate (%)', fontsize=14)
plt.ylim(0, 0.1)  # Setting the y-axis limit from 0 to 10% (0.1 as 10% failure)

# Show the value on top of the bars
for i, value in enumerate(failure_rate_data.values):
    plt.text(i, value + 0.001, f'{value*100:.2f}%', ha='center', va='bottom', fontsize=12)

# Optionally adjust x-axis labels for readability
plt.xticks(rotation=0)

# Show plot
plt.show()
```
