```python
import pandas as pd

df = pd.read_csv("data/shipping_data.csv")
df.head()
```

```python
# Shape and column names
print(df.shape)
print(df.columns)

# Data types and missing values
df.info()

# Descriptive statistics
df.describe()
```

```python
# Service distribution
df["Service"].value_counts()

# Delivery success rate
df["Was package delivered"].value_counts(normalize=True)

# Cross-tab of Service vs Delivery
pd.crosstab(df["Service"], df["Was package delivered"], normalize='index')
```

```python
# Missing values
df.isnull().sum()

# Duplicates
df.duplicated().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")
```

```python
sns.countplot(data=df, x="Service")
plt.title("Count of Orders by Shipping Service")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(
    data=df,
    x="Service",
    y="Was package delivered",
    estimator=lambda x: sum(x)/len(x)
)
plt.title("Delivery Success Rate by Service")
plt.ylabel("Success Rate")
plt.show()
```

```python
sns.boxplot(data=df, x="Service", y="Cost to deliver")
plt.title("Delivery Cost Distribution by Service")
plt.show()
```

```python
sns.histplot(data=df, x="Was package delivered", hue="Service", multiple="dodge", bins=2)
plt.title("Delivery Success by Service")
plt.xlabel("Was package delivered (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(df['Service'], df['Was package delivered'])
chi2, p_value, _, _ = chi2_contingency(contingency_table)

# Calculate average cost by service
avg_ground = df[df['Service'] == "UPS Ground"]['Cost to deliver'].mean()
avg_surepost = df[df['Service'] == "UPS SurePost"]['Cost to deliver'].mean()
cost_savings = (avg_ground - avg_surepost) * len(df)

# Print results
print("=== Contingency Table ===\n", contingency_table, "\n")
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Average Cost - UPS Ground: ${avg_ground:.2f}")
print(f"Average Cost - UPS SurePost: ${avg_surepost:.2f}")
print(f"Estimated Cost Savings for {len(df)} shipments: ${cost_savings:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate success rates
success_rates = df[df['Was package delivered'] == 1]['Service'].value_counts(normalize=True) * 100

# Plot
plt.figure(figsize=(8, 4))
sns.barplot(x=success_rates.values, y=success_rates.index, palette='viridis')
plt.xlabel("Delivery Success Rate (%)")
plt.title("Delivery Success Rate by Shipping Service")
plt.xlim(0, 100)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculate average cost per service
avg_costs = df.groupby('Service')['Cost to deliver'].mean()

# Plot
plt.figure(figsize=(8, 4))
sns.barplot(x=avg_costs.values, y=avg_costs.index, palette='coolwarm')
plt.xlabel("Average Delivery Cost ($)")
plt.title("Average Cost by Shipping Service")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Custom colors
colors = ['#FF9B00', '#0077B6']  # Replace with your palette

# Success rate
success_rates = df[df['Was package delivered'] == 1]['Service'].value_counts(normalize=True) * 100

# Plot
fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(success_rates.index, success_rates.values, color=colors)
ax.set_xlabel("Delivery Success Rate (%)")
ax.set_title("Delivery Success Rate by Shipping Service")
ax.set_xlim(0, 100)
ax.grid(axis='x', linestyle='--', alpha=0.6)
ax.bar_label(bars, fmt='%.2f%%', padding=3)
plt.tight_layout()
plt.show()
```

```python
# Average cost
avg_costs = df.groupby('Service')['Cost to deliver'].mean()

# Plot
fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(avg_costs.index, avg_costs.values, color=colors)
ax.set_xlabel("Average Delivery Cost ($)")
ax.set_title("Average Cost by Shipping Service")
ax.grid(axis='x', linestyle='--', alpha=0.6)
ax.bar_label(bars, fmt='$%.2f', padding=3)
plt.tight_layout()
plt.show()
```
